# Recommending model

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
final["index"]=final.index
```

```python
def get_title(i):
    return final[final.index==i]["recipe_name"].values[0]
def get_index(r):
    return final[final.recipe_name==r]["index"].values[0]
def get_recipe_id(i):
    return final[final.index==i]["recipe_id"].values[0]
```

```python
features=["ingredients","calories","fat","carbohydrates","protein","cholesterol","sodium","fiber","rating_x","number of rating"]
```

```python
def combine(row):
        return row["ingredients"]+" "+str(row["calories"])+" "+str(row["fat"])+" "+str(row["carbohydrates"])+" "+str(row["protein"])+" "+str(row["cholesterol"])+" "+str(row["sodium"])+" "+str(row["fiber"])+" "+str(row["rating_x"])+" "+str(row["number of rating"])
```

```python
final["combine_feature"]=final.apply(combine,axis=1)
```

```python
cv=CountVectorizer()
```

```python
count_matrix=cv.fit_transform(final["combine_feature"])
```

```python
cosine=cosine_similarity(count_matrix)
```

```python
cosine
```

```python
import urllib
import matplotlib.pyplot as plt
import random
```

```python
random.randint(0,50)
```

```python
def recommend(dish,ans):
    dish_index=get_index(dish)
    similar_dish=list(enumerate(cosine[dish_index]))
    recommend_in_order=sorted(similar_dish,key=lambda x:x[1],reverse=True)
    t=0
    #print("Here are the list of foods that we recommend based on your interest:")
    n=1
    for dish in recommend_in_order:
        print('\033[1m' +"{}.  ".format(n)+ get_title(dish[0]).upper() + '\033[0m')
        # create a file-like object from the url
        n+=1
        f = urllib.request.urlopen(final["image_url"][dish[0]])

        # read the image file in a numpy array
        a = plt.imread(f,format="jpg")
        plt.axis(False)
        plt.imshow(a)
        plt.show()
        if ans==1:
            print(final.iloc[dish[0]]["ingredients"])
        t=t+1
        if t>15:
            break
```

```python
def name(dish):
    dish_index=get_index(dish)
    similar_dish=list(enumerate(cosine[dish_index]))
    recommend_in_order=sorted(similar_dish,key=lambda x:x[1],reverse=True)
    return recommend_in_order
```

```python
# this function takes the userid and gives the name of a dish which that user likes the most.
def info(i_d):
    user=x[x["user_id"]==i_d]
    best_dish=pd.concat([user[user["rating"]==5],user[user["rating"]==4]],axis=0)
    nutri=["calories","fat","carbohydrates","protein","cholesterol","sodium","fiber"]
    l=[]
    for i in range(6,13):
        t=pd.to_numeric(best_dish.iloc[:,i]).sum()
        l.append(t)
    fav=nutri[l.index(max(l))]
    fav_dish=best_dish[best_dish[fav]==best_dish[fav].max()]["recipe_name"]
    return fav_dish
```
